```python
import pandas as pd
```

```python
path_mirna_clinical = r'C:\Users\Administrator\Desktop\LUAD_mirna_clinical.tsv'
path_rna_clinical = r'C:\Users\Administrator\Desktop\LUAD_rna_clinical.tsv'
```

```python
mirna_clinical_df = pd.read_csv(path_mirna_clinical, sep='\t')
mirna_ids = mirna_clinical_df['cases.submitter_id'].unique()
rna_clinical_df = pd.read_csv(path_rna_clinical, sep='\t')
rna_ids = rna_clinical_df['cases.submitter_id'].unique()
```

```python
common_ids = list(set(mirna_ids) & set(rna_ids))
```

```python
print("Common sample IDs:", common_ids)
```

```python
common_ids_df = pd.DataFrame(common_ids, columns=['common_sample_ids'])
common_ids_df.to_csv(r'C:\Users\Administrator\Desktop\LUAD_common_sample_ids.tsv', sep='\t', index=False)
```
